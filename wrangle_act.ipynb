### Twitter Archive Enhanced Table

#### Define: Change the data types of the tweet_id to strings and timestamps column to datetime

```python
twitter_archive_enhanced_clean.tweet_id = twitter_archive_enhanced_clean.tweet_id.astype(str)
twitter_archive_enhanced_clean.timestamp = pd.to_datetime(twitter_archive_enhanced_clean.timestamp, yearfirst=True)
twitter_archive_enhanced_clean.rating_numerator = twitter_archive_enhanced_clean.rating_numerator.astype(float)
twitter_archive_enhanced_clean.rating_denominator = twitter_archive_enhanced_clean.rating_denominator.astype(float)
```

#### Define
Separating Date and Time on the timestamp column then dropping the timestamp column

#### Code

```python
#Extracting date and time
twitter_archive_enhanced_clean['date'] = twitter_archive_enhanced_clean['timestamp'].apply(lambda x: x.strftime('%d-%m-%Y'))
twitter_archive_enhanced_clean['time'] = twitter_archive_enhanced_clean['timestamp'].apply(lambda x: x.strftime('%H:%M:%S'))

#changing datatype of the date column to datetime
twitter_archive_enhanced_clean.date = pd.to_datetime(twitter_archive_enhanced_clean.date, dayfirst = True)

# Dropping the timestamp column
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean.drop('timestamp', axis = 1) 

```

1.) Renaming Columns to more intuitive names
2.) Converting prediction confidence level to percentages
3.) Removing underscores and replacing with spaces  in the p1,p2,p3 prediction columns in the image_predictions table
4.) Changing the predictions column to sentence case



#### Code

```python
#Renaming columns for Twitter_Archive_enhanced table
image_predictions_clean = image_predictions_clean.rename({'jpg_url': 'image_url', 'p1':'prediction',
                                                        'p1_conf': 'confidence_level_percent', 'p1_dog': 'breed_predicted'},
                                                       axis = 'columns')


#Converting Confidence digits to percentages
image_predictions_clean.confidence_level_percent = image_predictions_clean.confidence_level_percent.apply(lambda x: round(x*100, 2))

#replacing the underscores in the prediction column with spaces
image_predictions_clean.prediction = image_predictions_clean.prediction.str.replace('_', ' ')

# #changing the predictions column to sentence case
image_predictions_clean.prediction = image_predictions_clean.prediction.str.title()
```

### Cleaning Tidiness

#### Define

#### Creating a column called dog_stage and melting the doggo,floofer,pupper,puppo columns

#### Code

```python
#Creating the dog_stage type column
twitter_archive_enhanced_clean['dog_stage'] = twitter_archive_enhanced_clean.text.str.extract('(doggo | floofer | pupper | puppo)', expand = True)

#dropping the doggo,floofer,pupper,puppo columns
twitter_archive_enhanced_clean = twitter_archive_enhanced_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis = 1)

#changing the datatype of dog_type to category
twitter_archive_enhanced_clean.dog_stage = twitter_archive_enhanced_clean.dog_stage.astype('category')

```

### Insights:

2.Comparing the percentage prediction confidence for the top 100 correctly predicted dog breeds and the top 100 incorrectly predicted dog breeds

3.Checking the average of the retweet_count, favorite_count, rating numerator and rating denominator

```python
#Creating datasets to for the dogs whose predictions were correct and the ones that were false

correct_predictions = twitter_archive_enhanced_clean[twitter_archive_enhanced_clean.breed_predicted == True]
incorrect_predictions = twitter_archive_enhanced_clean[twitter_archive_enhanced_clean.breed_predicted == False]
```

```python
#Checking the percentage success of the prediction programme
percent_prediction_success = (len(correct_predictions)/len(twitter_archive_enhanced_clean))*100
print(percent_prediction_success)
```

#### The prediction success of the neural network is 75%

#### Calculating the average percentage prediction confidence for the dogs that were correctly predicted and the ones that were incorrectly predicted

```python
# Calculating the average  percent prediction confidence for the top 100 correctly predicted dog breeds
(correct_predictions.groupby('prediction')['confidence_level_percent'].mean().sort_values(ascending = False).head(100)).mean()
```

```python
# Calculating the average percent prediction confidence for the top 100 incorrectly predicted dog breeds
(incorrect_predictions.groupby('prediction')['confidence_level_percent'].mean().sort_values(ascending = False).head(100)).mean()
```

#### The top 100 incorrectly predicted dog breeds had an average percent prediction confidence of 81% and the The top 100 correctly predicted dog breeds had an average percent prediction confidence of 58%. Incorrectly predicted breeds had a significant higher confidence level of prediction than the correctly predicted breeds.

#### Summarizing the average of the retweet_count, favorite_count, rating numerator of the top 20 dogs with the highest favorite count

```python
dog_perceptions = correct_predictions.groupby('prediction')['favorite_count','retweet_count'].mean()
dog_perceptions.sort_values(by = 'favorite_count', ascending = False).head(20)
```
